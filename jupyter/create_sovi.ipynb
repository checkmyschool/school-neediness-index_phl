# % Female

```python
from sklearn.preprocessing import LabelEncoder
discreteCoder_X = LabelEncoder()
```

```python
comb_values = comb.values
#comb_values
```

```python
#Make the SEX columns binary
comb_values[:,35] = discreteCoder_X.fit_transform(comb_values[:,35])
```

```python
comb2 = pd.DataFrame(comb_values)
comb2.columns = comb.columns
```

```python
num_female = comb2.groupby(['GEOLEV2'])['SEX'].sum()
#num_female
```

```python
sovi['sovi1'] = sovi['num_female']/sovi['num_persons']
#sovi
```

# % Non-Secondary Graduate Females Aged (20-39)

```python
comb2['age20_39'] = np.nan
comb2['fem20_39'] = np.nan
comb2['fem20_39_nodeg'] = np.nan
#comb2.head()
```

```python
#create binary age 20-39 column based on the age column
comb2.loc[((comb2['AGE'] >= 20) & (comb2['AGE'] <= 39)), 'age20_39'] = 1
comb2.loc[comb2['AGE'] < 20, 'age20_39'] = 0
comb2.loc[comb2['AGE'] > 39, 'age20_39'] = 0
```

```python
#create binary females aged 20-39 column based on the age20-39 and SEX column
comb2.loc[((comb2['age20_39'] == 1) & (comb2['SEX'] == 1)), 'fem20_39'] = 1
comb2.loc[(comb2['SEX'] == 0), 'fem20_39'] = 0
comb2.loc[(comb2['age20_39'] == 0), 'fem20_39'] = 0
```

```python
#create binary females aged 20-39 with no secondary degree column based on the age20-39 and EDUCPH column
comb2.loc[((comb2['fem20_39'] == 1) & (comb['EDUCPH'] <= 43)), 'fem20_39_nodeg'] = 1
comb2.loc[(comb2['fem20_39'] == 0), 'fem20_39_nodeg'] = 0
```

```python
#comb2.head(20)
```

```python
num_fem20_39_nodeg = comb2.groupby(['GEOLEV2'])['fem20_39_nodeg'].sum()
#num_fem20_39_nodeg
```

```python
sovi['num_fem20_39_nodeg'] = num_fem20_39_nodeg.tolist()
#sovi.head()
```

```python
sovi['sovi5'] = sovi['num_fem20_39_nodeg']/sovi['num_persons']
#sovi.head()
```
